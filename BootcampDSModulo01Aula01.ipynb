## Análise dos dados do DataSUS a partir de 2008

# Aula 01

```
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```
dados = pd.read_csv("/content/A151346189_28_143_208.csv", encoding="ISO-8859-1", 
                    skiprows=3, sep=";", skipfooter=12,
                    thousands=".", decimal=",",
                    na_values='-', engine="python")
dados
```

```
pd.options.display.float_format = "{:.2f}".format
```

```
dados.mean()
```

```
dados["2008/Ago"].mean()
```

```
dados.plot(x="Unidade da Federação", y="2008/Ago")
```

```
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,6))
```

```
# Desafio: Alterar o títilo do gráfico para que fique mais compreensível
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos em procedimentos hospitalares pelo SUS por Unidade da Federação")
plt.show()
```

```
# Desafio: Fazer o mesmo gráfico acima, porém com os dados do mês mais recente
ax = dados.sort_values("2020/Ago", ascending=False).plot(x="Unidade da Federação", y="2020/Ago", kind="bar", figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor gasto em procedimentos hospitalares pelo SUS por Unidade da Federação")
plt.show()
```
